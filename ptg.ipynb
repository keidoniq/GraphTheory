```python
def visualize_tree(tree):
    def add_nodes_edges(tree, dot=None):
        # Create Digraph object
        if dot is None:
            dot = Digraph()
            dot.node(name=str(tree), label=str(tree.val))

        # Add nodes
        if tree.left:
            dot.node(name=str(tree.left) ,label=str(tree.left.val))
            dot.edge(str(tree), str(tree.left))
            dot = add_nodes_edges(tree.left, dot=dot)

        if tree.right:
            dot.node(name=str(tree.right) ,label=str(tree.right.val))
            dot.edge(str(tree), str(tree.right))
            dot = add_nodes_edges(tree.right, dot=dot)

        return dot

    # Add nodes recursively and create a list of edges
    dot = add_nodes_edges(tree)

    # Visualize the graph
    display(dot)

    return dot
```

# Задача 13.
Найти максимальное паросочетание в чередующемся дереве.

## Решение

```python
dot = Digraph(graph_attr={'rankdir':'RL', 'ranksep': "equally"},
              node_attr={'shape' : 'circle'},
              edge_attr={'dir':'none'})

nodes = [chr(i) for i in range(ord('a'),ord('i'))]
nodes.append('k')
nodes.append('p')
nodes.append('s')

edges = ['ab', 'bc', 'dc', 'ed', 'cf', 'kh', 'fk', 'kp', 'ps', 'hg']

for nd in nodes:
    dot.node(nd)
dot.edges(edges)
dot
```

*Чередующееся дерево* - дерево, вершины которого разбиты на два п/мн-ва -- внутренние и внешние -- при чем:
*   каждое ребро соединяет внутреннюю вершину с внешней,
*   степень каждой из внутренних вершин = 2,
*   ???????



**S1.** Получим разбиение вершин на внутренние, выделенные цветом, и внешние:


```python
inner_nodes = ['p','h','f','d','b']
outer_nodes = [nd for nd in nodes if nd not in inner_nodes]

for nd in inner_nodes:
    dot.node(nd,_attributes={'fillcolor':'yellow','style':'filled'})#'color':'green'})
dot
```

**S2.** Выберем в качестве ведущей вершины $x_0$ произвольную вершину из мн-ва внешних вершин
$x_0 = c$ и положим ей в соот-вие метку $d_0 = 0$.
В качестве меток для остальных вершин примем длину кратчайшей цепи от данной вершины к вершине $x_0$. Получим:

```python
distance = {
    'a': 2,    'b': 1,    'c': 0,    'd': 1,    'e': 2,
    'f': 1,    'g': 4,    'h': 3,    'k': 2,    'p': 3,    's': 4
}

for nd in nodes:
    if nd in inner_nodes:
        dot.node(nd,_attributes={'fillcolor':'yellow','style':'filled','xlabel':str(distance[nd])})
    else:
        dot.node(nd,_attributes={'xlabel':str(distance[nd])})
dot
```

**S3.** В максимальное паросочетание включим такие ребра $( x_i^{внутр}, x_j^{внешн})$, для концевых вершин которых будет верно: $d_j = d_i + 1$


То есть:

```python
max_pair = ['hg', 'ps', 'fk', 'ab', 'ed']
```

```python
for ed in max_pair:
    dot.edge(ed[0],ed[1],_attributes={'color':'red'})
dot
```

Таким образом, максимальное паросочетание данного графа включает в себя следующие ребра:

```python
dot2 = Digraph(graph_attr={'rankdir':'RL', 'ranksep': "equally"},
              node_attr={'shape' : 'circle'},
              edge_attr={'dir':'none'})

for nd in nodes:
    dot2.node(nd)

edges = ['ab', 'bc', 'dc', 'ed', 'cf', 'kh', 'fk', 'kp', 'ps', 'hg']
for ed in edges:
    if ed in max_pair:
        dot2.edge(ed[0],ed[1],_attributes={'color':'red'})
    else:
        dot2.edge(ed[0],ed[1])

dot2
```
